```python
import numpy as np
import pandas as pd
import csv
from datetime import datetime
import multiprocessing
# import pdb; pdb.set_trace()
filename='/Users/dengweishu/Desktop/MC.txt'
data=pd.read_csv(filename)
series=data['avgLogLiso']

 
def BM(batch_size,batch_jump,length=len(series)):
        num=int(float(len(series))/float(batch_size))
        i=0
        T=list()
        while i<len(series) :
                hold_list=series[i:i+batch_size]
                batch_mean=np.mean(hold_list)
                T.append(batch_mean)
                i=i+batch_jump
        variance=np.var(series[:i]) #effective series variance
        IAC=batch_size*np.var(T)/variance
        return IAC



hold=int(len(series)**0.666666)
x=list()
y=list()
for batch_size in range (1,hold):
         
        x.append(batch_size)
         
        y.append(BM(batch_size,batch_size))
            

                 
 

```

```python
plt.plot(x,y)
plt.xlabel('batch size')
plt.ylabel('IAC')
iac=[34]*len(x)
plt.plot(x,iac)
plt.title('MC series')
```
